```python
# --- Standard Library ---
import os
import time

# --- Third-Party Libraries ---
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister

# --- Local Imports ---
from qiskit_quantum_simulator import QiskitQuantumSimulator
```

```python
qc = QuantumCircuit()

q = QuantumRegister(2, 'q')
c = ClassicalRegister(2, 'c')

qc.add_register(q)
qc.add_register(c)

qc.h(q[0])
qc.cx(q[0], q[1])
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])

ibm_simulator = QiskitQuantumSimulator(ibm_backend='ibm_brisbane')
job = ibm_simulator.run(qc)
job_result = job.result()

print(job_result[0].data.c.get_counts())
ibm_simulator.transpile_circuit(qc).draw('mpl')
```
